```python
sns.pairplot(df.drop(columns=["Date"]))
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop(columns=["Date"]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x="CO2_ppm", y="Product_Demand", data=df, hue="Humidity_percent", size="Temperature_C", sizes=(50, 200))
plt.title("CO₂ vs Product Demand")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df.drop(columns=["Date"]))
plt.title("Boxplot of All Features")
plt.show()

```

```python
sns.jointplot(x="Temperature_C", y="Product_Demand", data=df, kind="hex", cmap="viridis")
plt.show()
```

```python
sns.histplot(df["Product_Demand"], bins=10, kde=True, color="purple")
plt.title("Distribution of Product Demand")
plt.xlabel("Product Demand")
plt.ylabel("Frequency")
plt.show()
```

 ## MATEPLOTLIB VISUALIZATION USING PLOT BASICS

```python
import matplotlib.pyplot as plt
## matplotlib for viszualization
x=[1,2,3,4,5]
y=[2,4,6,8,10]      
z=[5,4,8,9,4]
plt.plot(x,y)
plt.title("plots data")

plt.show()
```

## sub plot

```python
plt.subplot(1, 2, 1)   # (rows, cols, position)
plt.plot(x, y, marker="o", color="b")
plt.title("X vs Y")

plt.subplot(1, 2, 2)   # 2nd plot
plt.plot(x, z, marker="s", color="r")
plt.title("X vs Z")


plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sizes = [40, 25, 20, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]
colors = ["red", "yellow", "pink", "brown"]

# Pie chart
plt.pie(
    sizes, 
    labels=labels, 
    colors=colors, 
    autopct="%1.1f%%",   # show percentages
    startangle=90,       # rotate start angle
    shadow=True,         # add shadow
    explode=(0.1, 0, 0, 0)  # "explode" (separate) 1st slice
)

plt.title("Fruit Distribution")
plt.show()
```
